# Regression

```python
import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
import geopandas as gpd
import seaborn as sns
sns.set()
```

### Load dataset ANJUKE

```python
anjuke = pd.read_csv('UTSEUS-anjuke-real-estate(1).csv')
```

```python
anjuke.describe()
```

```python
# Filter outliers using z-score on longitude and latitude
anjuke_lon_z = (anjuke.longitude - anjuke.longitude.mean()) / anjuke.longitude.std()
anjuke_lat_z = (anjuke.latitude - anjuke.latitude.mean()) / anjuke.latitude.std()

an_df = anjuke[(anjuke_lon_z < 3) & (anjuke_lat_z < 3)]
an_df.describe()
```

```python
# Create GeoDataFrame from filtered data
an_gdf = gpd.GeoDataFrame(
    an_df,
    geometry=gpd.points_from_xy(
        an_df.longitude,
        an_df.latitude),
    crs=4326)

an_gdf.plot(markersize=1, color='b', figsize=(10, 10))
```

```python
# Load the Shanghai boundary (adcode: 310000) from Aliyun DataV GeoAtlas
aliyun_shanghai_url = "https://geo.datav.aliyun.com/areas_v3/bound/310000.json"
shanghai_boundary = gpd.read_file(aliyun_shanghai_url)

shanghai_boundary.plot()
```

```python
shanghai_boundary.crs
```

```python
# Clip real estate points to Shanghai boundary
ansh_gdf = gpd.clip(an_gdf, shanghai_boundary)

# Plot to verify the results
fig, ax = plt.subplots(figsize=(10, 10))
shanghai_boundary.plot(ax=ax, color='white', edgecolor='black', linewidth=0.5)
ansh_gdf.plot(ax=ax, color='red', markersize=2, alpha=0.5, label="Real Estate")
ax.set_title("Shanghai Real Estate Points within Aliyun Boundary")
plt.show()
```

### Load dataset POI

```python
poi_df = pd.read_csv('shanghaiPOI2017.csv', delimiter=';', decimal=',', low_memory=False)
```

```python
# Inspect rows with spaces in TYPE field
poi_df[poi_df.TYPE.str.contains(' ')].loc[:, ['TYPE', 'GBCODE']]
```

```python
# Filter subway stations (GBCODE = 200402)
subway = poi_df[poi_df.GBCODE == 200402]

subway_gdf = gpd.GeoDataFrame(
    subway,
    geometry=gpd.points_from_xy(
        subway.GPSX,
        subway.GPSY),
    crs=4326)
```

```python
# Plot real estate and subway stations together
fig, ax = plt.subplots(figsize=(10, 10))
shanghai_boundary.plot(ax=ax, color='white', edgecolor='black', linewidth=0.5)
ansh_gdf.plot(ax=ax, color='red', markersize=1, alpha=0.2, label="Real Estate")
subway_gdf.plot(ax=ax, color='green', markersize=1, alpha=1, label="Subway")
ax.set_title("Shanghai Real Estate Points within Aliyun Boundary")
plt.legend()
plt.show()
```

### Nearest distance

```python
# Reproject to a metric CRS for distance calculations
ansh_gdf = ansh_gdf.to_crs(4576)
subway_gdf = subway_gdf.to_crs(4576)

ansh_gdf.plot()
```

```python
# Spatial join: find nearest subway station for each real estate point
joined_gdf = gpd.sjoin_nearest(ansh_gdf, subway_gdf, distance_col="distances")
```

```python
# Keep only relevant columns
to_be_deleted_cols = [c for c in joined_gdf.columns if c not in ['id', 'onesquaremeter', 'geometry', 'distances']]
joined_gdf = joined_gdf.drop(columns=to_be_deleted_cols)
joined_gdf.info()
```

```python
# Plot price per sqm vs distance to nearest subway
sns.regplot(joined_gdf, x='distances', y='onesquaremeter')
```

```python
joined_gdf.describe()
```

```python
# Z-score filtering on price per sqm to remove extreme outliers
joined_gdf['z'] = (joined_gdf.onesquaremeter - joined_gdf.onesquaremeter.mean()) / joined_gdf.onesquaremeter.std()
```

```python
# Regplot with outlier filtering
sns.regplot(
    joined_gdf[(joined_gdf.z < 5) & (joined_gdf.distances < 35_000)],
    x='distances',
    y='onesquaremeter'
)
```
